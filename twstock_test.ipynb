```python
import pandas as pd
```

```python
l = pd.read_csv('./data/new_stocks.csv',index_col=0)
```

```python
symbols = [str(i).zfill(4) for i in l['代號'].to_list()]
```

```python
symbols
```

```python
info_buy = list()
info_sell = list()
```

```python
for s in symbols:
    try:
        stock = Stock(s)
    except:
        continue
        
    try:
        price = stock.price
    except:
        continue
    
    if None in price:
        continue
    
    bfp = BestFourPoint(stock)
    try:
        a = bfp.best_four_point()
    except:
        continue
    if a:
        if a[0] == True:
            info_buy.append(a)
        else:
            info_sell.append(a)
    
    
```
